```python
counts = df['no_of_dependents'].value_counts()
```

```python
df['loan_status'].value_counts().plot(kind='bar')
plt.title("Loan Approval Count")
plt.xlabel("Loan Status")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(df['income_annum'], bins=20)
plt.title("Income Distribution")
plt.xlabel("Annual Income")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(df['loan_amount'])
plt.title("Loan Amount Outliers")
plt.show()
```

```python
pd.crosstab(df['self_employed'], df['loan_status']).plot(kind='bar')
plt.title("Education vs Loan Status")
plt.show()
```

```python
plt.scatter(df['income_annum'], df['loan_amount'])
plt.xlabel("Income")
plt.ylabel("Loan Amount")
plt.title("Income vs Loan Amount")
plt.show()
```

```python
pd.crosstab(df['loan_status'], df['cibil_score'] > 700).plot(kind='bar')
plt.title("Loan Status vs High CIBIL Score")
plt.show()
```
